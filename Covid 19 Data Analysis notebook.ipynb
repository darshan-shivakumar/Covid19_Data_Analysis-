# COVID-19 Data Analysis using Python

### Import the modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing the covid confirmed cases dataset

```python
covid_data= pd.read_csv("covid19_Confirmed_dataset.csv")
covid_data.head()
```

```python
covid_data.shape
```

### Removing the useless columns 

```python
covid_data.drop(['Lat','Long'],axis=1,inplace=True)
covid_data.head()
```

```python
covid_agg = covid_data.groupby("Country/Region").sum()
covid_agg.head(10)
```

```python
covid_agg.shape
```

## Visualizing data related to particular countries

```python
covid_agg.loc['India']
```

```python
# Now lets plot it as a graph

covid_agg.loc['India'].plot()
```

### Now let's plot wrt Italy and spain 

```python
covid_agg.loc['India'].plot()
covid_agg.loc['China'].plot()
covid_agg.loc['Italy'].plot()
covid_agg.loc['Spain'].plot()
plt.legend()
```

## Calculating a good measure

```python
# Taking China as an example

covid_agg.loc['China'].plot()
```

```python
# Now let's plot for only first 3 days

covid_agg.loc['China'][:3].plot()
```

## As we can see the increase in cases each day, lets plot first derivative of curve to find the rate of maximum cases

```python
covid_agg.loc['China'].diff().plot()
```

```python
covid_agg.loc['China'].diff().max()
```

```python
# similarly for India
covid_agg.loc['India'].diff().plot()
```

## Now let us calculate highest one day record for all countries 

```python
# lets make a list of all countries

countries= list(covid_agg.index)
max_inf_rate=[]
for i in countries:
    max_inf_rate.append(covid_agg.loc[i].diff().max())
    
covid_agg['Maximum_infection_rate']= max_inf_rate
covid_agg.head()
```

## Now let's remove other columns

```python
# we create a new dataframe using pandas 

covid19_data= pd.DataFrame(covid_agg['Maximum_infection_rate'])
```

```python
covid19_data.head()
```

# Importing the Happiness report dataset

```python
happ_report=pd.read_csv('worldwide_happiness_report.csv')
happ_report.head(10)
```

### Let's remove the unwanted columns 

```python
columns=['Overall rank','Score','Generosity','Perceptions of corruption']

happ_report.drop(columns,axis=1,inplace=True)

happ_report.head(10)
```

```python
# Now let's make the country/region as index

happ_report.set_index('Country or region',inplace=True)

happ_report.head(10)
```

# Joining both confirmed cases and happiness report datasets.

```python
covid19_data.head(10)
```

```python
covid19_data.shape
```

```python
happ_report.head(10)
```

```python
happ_report.shape
```

```python
# We perform the join operation using inner join since both datasets don't have same no.of rows

covid_19 = covid19_data.join(happ_report,how='inner')

covid_19.head(15)
```

## Correlation

```python
# correlation gives the dependency of factors on each other

covid_19.corr()
```
